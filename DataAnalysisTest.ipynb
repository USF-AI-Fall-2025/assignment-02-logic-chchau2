```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("gallstone.csv")

df.info()
print(df["Gender"].value_counts())

#men have higher hemoglobin levels
sns.boxplot(x="Gender",y="Hemoglobin (HGB)", data=df)
plt.title("Hemoglobin by Gender Group")
plt.show()

#men have higher muscle mass
sns.boxplot(x="Gender", y="Muscle Mass (MM)", data=df)
plt.title("Muscle Mass by Gender")
plt.show()

#men have higher lean mass
sns.boxplot(x="Gender", y="Lean Mass (LM) (%)", data=df)
plt.title("Lean Mass by Gender")
plt.show()

#check distribution of glucose
sns.displot(df["Glucose"])
plt.show()
```
